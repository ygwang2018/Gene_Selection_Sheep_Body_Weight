```python
import numpy as np
import random
from sklearn.svm import SVR
def BPSO(X_train,Y_train,X_test,Y_test,T,N,v_max,w,c_1,c_2,s,c,Gamma,Epsilon):
    sample_num,feature_num=X_train.shape
    Sample_test_num,feature_num=X_test.shape
    Velocity=[]
    X=[]
    best_MSE=float("inf")
    best_active=[]
    p=s/feature_num
    clf=SVR(kernel='rbf',C=c,gamma=Gamma,epsilon=Epsilon)
    for i in range(N):
        veloci_indi=[random.uniform(-v_max,v_max) for num in range(feature_num)]
        Velocity.append(veloci_indi)
        lo_indi=[1 if random.uniform(0,1)<p else 0 for i in range(feature_num)]
        X.append(lo_indi)
    global_best=[0,float("inf")]
    local_best_lis=[[i,float("inf")] for i in range(N)]
    t=1
    while t<=T:
        Fitness=[]
        for each_fit in range(N):
            Active_set=[]
            for fea in range(feature_num):
                if X[each_fit][fea]==1:
                    Active_set.append(fea)
            clf.fit(X_train[:,Active_set],Y_train.ravel())
            Y_predict=clf.predict(X_test[:,Active_set]).reshape(Sample_test_num,1)
            MSE=np.sum((Y_predict-Y_test)**2)/Sample_test_num
            Fitness.append(MSE)
            print(MSE)
            Fitness.append(MSE)
            if MSE<best_MSE:
                best_MSE=MSE
                best_active=Active_set
            if MSE<local_best_lis[each_fit][1]:
                local_best_lis[each_fit][1]=MSE
                local_best_lis[each_fit][0]=X[each_fit]
            if MSE<global_best[1]:
                global_best[0]=X[each_fit]
                global_best[1]=MSE
        for ind_ in range(N):
            for fea in range(feature_num):
                r_1i=random.uniform(0,1)
                r_2i=random.uniform(0,1)
                Velocity[ind_][fea]=w*Velocity[ind_][fea]+c_1*r_1i*(local_best_lis[ind_][0][fea]-X[ind_][fea])+c_2*r_2i*(global_best[0][fea]-X[ind_][fea])
                sigmoid=1/(1+np.exp(-Velocity[ind_][fea]))
                if random.uniform(0,1)<sigmoid:
                    X[ind_][fea]=1
                else:
                    X[ind_][fea]=0
        t+=1
    return best_active,best_MSE
                    
                
        
        
        
        
```

```python
import pandas as pd
import numpy as np
data=pd.read_csv("HuSheepData.csv",encoding='gb18030')
Weaning_Weight=np.array(data)[5,1:].astype(np.float64).reshape(240,1)
Birth_Weight=np.array(data)[4,1:].astype(np.float64).reshape(240,1)
Six_Weight=np.array(data)[6,1:].astype(np.float64).reshape(240,1)
Featur_Sample=np.array(data)[8:,1:].astype(np.float64).T
X_train=Featur_Sample[:170,:]
X_test=Featur_Sample[170:,:]
Birth_Weight_train=Birth_Weight[:170]
Birth_Weight_test=Birth_Weight[170:]
Six_Weight_train=Six_Weight[:170]
Six_Weight_test=Six_Weight[170:]
Weaning_Weight_train=Weaning_Weight[:170]
Weaning_Weight_test=Weaning_Weight[170:]
```

# Birth Weight

```python
Birth_best_active,Birth_best_MSE=BPSO(X_train,Birth_Weight_train,X_test,Birth_Weight_test,30,20,6,0.7298,1.49618,1.49618,2000,0.1,0.01,0.00001)
```

```python
col=np.array(data)[:,0]
gene_name=col[8:]
gene=pd.DataFrame(gene_name[Birth_best_active])
gene.to_csv("best_gene_birth_BPSO.csv")
```
