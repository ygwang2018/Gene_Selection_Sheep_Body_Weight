```python
from Initia_Oper import Initialization
import numpy as np
from Crossover_Uniform import uniform_crossover_operator
from Mutation import Standard_Bit_Mutation
from sklearn.svm import SVR
from Fitness_pro_sele import FPS
import random
def GGA(X_train,Y_train,X_test,Y_test,T,N,P_mutation,P_cross,s,c,Gamma,Epsilon):
    t=1
    Population=[]
    sample_num_train,feature_num=X_train.shape
    sample_num_test,feature_num=X_test.shape
    clf=SVR(kernel='rbf',C=c,gamma=Gamma,epsilon=Epsilon)
    #best_SIC=float("inf")
    best_MSE=float("inf")
    best_active=[]
    Min_Fitness=[]
    Max_Fitness=[]
    Mean_Fitness=[]
    #初始化
    for indi in range(N):
        #Individual=Initialization(s,feature_num)
        #Individual=[random.randint(0,1) for i in range(feature_num)]
        Individual=[1 if random.uniform(0,1)<(s/feature_num) else 0 for i in range(feature_num)]
        Population.append(Individual)
    while t<=T:
        #calculate Fitness
        Fitness=[]
        for each_fit in range(N):
            Active_set=[]
            for fea in range(feature_num):
                if Population[each_fit][fea]==1:
                    Active_set.append(fea)
            clf.fit(X_train[:,Active_set],Y_train.ravel())
            Y_predict=clf.predict(X_test[:,Active_set]).reshape(sample_num_test,1)
            MSE=np.sum((Y_predict-Y_test)**2)/sample_num_test
            #SIC=sample_num_test*np.log(MSE)+len(Active_set)*np.log(feature_num)*np.log(np.log(sample_num_test))
            print(MSE)
            if MSE<best_MSE:
                #best_SIC=SIC
                best_active=Active_set
                best_MSE=MSE
            Fitness.append(MSE)
        Min_Fitness.append(np.min(Fitness))
        Max_Fitness.append(np.max(Fitness))
        Mean_Fitness.append(np.mean(Fitness))
        new_population=[]
        for i in range(int(N/2)):
        # Selection Operator
            Parent1=Population[FPS(Fitness,N)]
            Parent2=Population[FPS(Fitness,N)]
        #crossover Operator
            P_0=[P_cross for i in range(feature_num)]
            offspring1=uniform_crossover_operator(Parent1,Parent2,P_0)
            offspring2=uniform_crossover_operator(Parent1,Parent2,P_0)
        #mutation Operator
            P_m=[P_mutation for i in range(feature_num)]
            offspring1=Standard_Bit_Mutation(offspring1,P_m)
            offspring2=Standard_Bit_Mutation(offspring2,P_m)
            new_population.append(offspring1)
            new_population.append(offspring2)
            
        Population=new_population
        t+=1
        print(t)
    return best_active,best_MSE,Min_Fitness,Max_Fitness,Mean_Fitness
```

```python
import pandas as pd
import numpy as np
data=pd.read_csv("HuSheepData.csv",encoding='gb18030')
Weaning_Weight=np.array(data)[5,1:].astype(np.float64).reshape(240,1)
Birth_Weight=np.array(data)[4,1:].astype(np.float64).reshape(240,1)
Six_Weight=np.array(data)[6,1:].astype(np.float64).reshape(240,1)
Featur_Sample=np.array(data)[8:,1:].astype(np.float64).T
X_train=Featur_Sample[:170,:]
X_test=Featur_Sample[170:,:]
Birth_Weight_train=Birth_Weight[:170]
Birth_Weight_test=Birth_Weight[170:]
Six_Weight_train=Six_Weight[:170]
Six_Weight_test=Six_Weight[170:]
Weaning_Weight_train=Weaning_Weight[:170]
Weaning_Weight_test=Weaning_Weight[170:]
```
