```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.cbook as cbook
import matplotlib.ticker
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.gridspec import GridSpec
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
from mpl_toolkits.axes_grid1.inset_locator import mark_inset
% matplotlib inline
font1={'family': 'Times New Roman', 'weight': 'light', 'size': 16}#
font2={'family': 'Times New Roman', 'weight': 'bold', 'size': 16}#
fig = plt.figure(figsize=(25,20),facecolor="snow")
#fig.suptitle("Controlling spacing around and between subplots")

gs1 = GridSpec(20, 20)
ax1 = fig.add_subplot(gs1[0:5, 0:5],facecolor="whitesmoke")
ax2 = fig.add_subplot(gs1[0:5, 6:11],facecolor="whitesmoke")
ax3 = fig.add_subplot(gs1[6:11,0:5],facecolor="whitesmoke")
data=pd.read_csv("mean_fitness.csv")
arra_data=np.array(data)
propo_GA_birth=arra_data[0,1:21]
GA_birth=arra_data[1,1:]

propo_GA_six=arra_data[2,1:28]
GA_six=arra_data[3,1:]

propo_GA_weaning=arra_data[4,1:13]
GA_weaning=arra_data[5,1:]

generation=[i for i in range(1,31)]

#mean fitness,propos_GA
ax1.plot(generation[:20],propo_GA_birth,"r-",label="Proposed Method",lw=2)
#GA_birth
ax1.plot(generation,GA_birth,"g-",label="GA",lw=2)

ax1.xaxis.grid(True, which='major')
ax1.yaxis.grid(True, which='major')
ax1.set_xlabel("Number of Generation",fontsize=13)
ax1.set_title("Birth Weight",fontsize=13)
ax1.set_ylabel('Mean Fitness',fontsize=13)
ax1.tick_params(labelsize=13)
ax1.ticklabel_format(axis='both',style='sci')#sci
ax1.legend(bbox_to_anchor=(5.2,-0.3),prop=font1,shadow=True,frameon=False)

# Six
ax2.plot(generation[:27],propo_GA_six,"r-",label="Proposed Method",lw=2)
ax2.plot(generation,GA_six,"g-",label="GA",lw=2)

ax2.xaxis.grid(True, which='major')
ax2.yaxis.grid(True, which='major')
ax2.set_xlabel("Number of Generation",fontsize=13)
ax2.set_title("Six-Month Weight",fontsize=13)
ax2.set_ylabel('Mean Fitness',fontsize=13)
ax2.tick_params(labelsize=13)
ax2.ticklabel_format(axis='both',style='sci')#sci


ax3.plot(generation[:12],propo_GA_weaning,"r-",label="Proposed Method",lw=2)
ax3.plot(generation,GA_weaning,"g-",label="GA",lw=2)
ax3.xaxis.grid(True, which='major')
ax3.yaxis.grid(True, which='major')
ax3.set_xlabel("Number of Generation",fontsize=13)
ax3.set_title("Weaning Weight",fontsize=13)
ax3.set_ylabel('Mean Fitness',fontsize=13)
ax3.tick_params(labelsize=13)
ax3.ticklabel_format(axis='both',style='sci')#sci
ax1.legend(bbox_to_anchor=(1.8,-0.6),prop=font1,shadow=True,frameon=False)
#plt.show()
plt.savefig("Fitness_mean.eps",format='eps',dpi=1000, bbox_inches='tight')
```
