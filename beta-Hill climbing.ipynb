```python
import random
import numpy as np
from time import *
import copy
from sklearn.svm import SVR 
def Binary_beta_Hill_climbing(X_train,Y_train,X_test,Y_test,T,N,eta,theta,s,c,Gamma,Epsilon):
    t=1
    sample_num,feature_num=X_train.shape
    Sample_test_num,feature_num=X_test.shape
    best_active=[]
    best_MSE=float("inf")
    Population=[]
    p=s/feature_num
    clf=SVR(kernel='rbf',C=c,gamma=Gamma,epsilon=Epsilon)
    #初始化
    for indi_ in range(N):
        individual=[1 if random.uniform(0,1)<p else 0 for i in range(feature_num)]
        Population.append(individual)
    Fitness=[]
    for each_fit in range(N):
        Active_set=[]
        for fea in range(feature_num):
            if Population[each_fit][fea]==1:
                Active_set.append(fea)
        clf.fit(X_train[:,Active_set],Y_train.ravel())
        Y_predict=clf.predict(X_test[:,Active_set]).reshape(Sample_test_num,1)
        MSE=np.sum((Y_predict-Y_test)**2)/Sample_test_num
        Fitness.append(MSE)
        if MSE<best_MSE:
            best_MSE=MSE
            best_active=Active_set
        Fitness.append(MSE)
    while t<=T:
       
        #print(t,mean_Fitness)
        for indivi_ide in range(N):
            new_individual=copy.deepcopy(Population[indivi_ide])
            #print(new_individual)
            #N-operator
            sele_fea=random.randint(0,feature_num-1)
            new_individual[sele_fea]=new_individual[sele_fea]+random.uniform(-eta,eta)
            for fea in range(feature_num):
                #beta-operator
                if random.uniform(0,1)<=theta:
                    new_individual[fea]=random.randint(0,1)   
                Sigmoid=1/(1+np.exp(-new_individual[fea]))
                if random.uniform(0,1)<Sigmoid:
                    new_individual[fea]=1
                else:
                    new_individual[fea]=0
            Active_set=[]
            for fea in range(feature_num):
                if new_individual[fea]==1:
                    Active_set.append(fea)
            clf.fit(X_train[:,Active_set],Y_train.ravel())
            Y_predict=clf.predict(X_test[:,Active_set]).reshape(Sample_test_num,1)
            MSE=np.sum((Y_predict-Y_test)**2)/Sample_test_num
            if MSE<best_MSE:
                best_MSE=MSE
                best_active=Active_set
       
            if MSE<=Fitness[indivi_ide]:
                Fitness[indivi_ide]=MSE
                Population[indivi_ide]=new_individual
            #print(indivi_ide)
            #print(best_SIC)
        t=t+1
        print(t)
    return best_active,best_MSE
                    
                
```

```python
import pandas as pd
import numpy as np
data=pd.read_csv("HuSheepData.csv",encoding='gb18030')
Weaning_Weight=np.array(data)[5,1:].astype(np.float64).reshape(240,1)
Birth_Weight=np.array(data)[4,1:].astype(np.float64).reshape(240,1)
Six_Weight=np.array(data)[6,1:].astype(np.float64).reshape(240,1)
Featur_Sample=np.array(data)[8:,1:].astype(np.float64).T
X_train=Featur_Sample[:170,:]
X_test=Featur_Sample[170:,:]
Birth_Weight_train=Birth_Weight[:170]
Birth_Weight_test=Birth_Weight[170:]
Six_Weight_train=Six_Weight[:170]
Six_Weight_test=Six_Weight[170:]
Weaning_Weight_train=Weaning_Weight[:170]
Weaning_Weight_test=Weaning_Weight[170:]
```

# Birth Weight

```python
Birth_best_active,Birth_best_MSE=Binary_beta_Hill_climbing(X_train,Birth_Weight_train,X_test,Birth_Weight_test,30,20,0.9,0.5,2000,0.1,0.01,0.00001)
```
